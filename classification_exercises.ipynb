```python
# Make a function named get_iris_data that returns the data from the iris_db on the codeup data science database as a pandas data frame. 
# The returned data frame should include the actual name of the species in addition to the species_ids. Obtain your data from the Codeup Data Science Database.
from env import user, host, password
import pandas as pd
import numpy as np
from pydataset import data
import os

def get_iris_data(
    db,
    user=user,
    password=password,
    host=host):
    '''
    grab data will query data from a specified positional argument (string literal)
    schema from an assumed user, password, and host provided
    that they were imported from an env
    
    return: a pandas dataframe
    '''
    query = '''SELECT * FROM measurements JOIN species USING (species_id)'''
    connection = f"mysql+pymysql://{user}:{password}@{host}/{db}"
    df = pd.read_sql(query, connection)
    return df
get_iris_data('iris_db')
```

```python
# Make a function named get_telco_data that returns the data from the telco_churn database in SQL. 
# In your SQL, be sure to join contract_types, internet_service_types, payment_types tables with the customers table, 
# so that the resulting dataframe contains all the contract, payment, and internet service options. Obtain your data from the Codeup Data Science Database.
def get_telco_data(
    db,
    user=user,
    password=password,
    host=host):
    '''
    grab data will query data from a specified positional argument (string literal)
    schema from an assumed user, password, and host provided
    that they were imported from an env
    
    return: a pandas dataframe
    '''
    query = '''SELECT * FROM contract_types JOIN internet_service_types JOIN payment_types'''
    connection = f"mysql+pymysql://{user}:{password}@{host}/{db}"
    df = pd.read_sql(query, connection)
    return df
get_telco_data('telco_churn')
```

```python
# Write a function to split your data into train, test and validate datasets. Add this function to prepare.py.
def train_val_test(df, strat, seed = 123):
    train, val_test = train_test_split(df, train_size = 0.8, random_state = seed, stratify = df[strat])
    val, test = train_test_split(val_test, train_size = 0.5, random_state = seed, stratify = val_test[strat])
    return train, val, test
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Univariate Stats

# For each measurement type (quantitative variable): create a histogram, boxplot, & compute descriptive statistics (using .describe()).
quan, cat = [], []
for col in train_iris:
    if train_iris[col].dtype=='float64':
        quan.append(col)
for col in quan:
    plt.hist(train_iris[col])
    plt.title(f'Distribution of {col}')
    plt.show()
    sns.boxplot(data=train_iris, x=col)
    plt.show()
    print(train_iris[col].describe())
```

```python
# For each species (categorical variable): create a frequency table and a bar plot of those frequencies.
quan, cat = [], []
for col in train_iris:
    if train_iris[col].dtype=='O':
        cat.append(col)
for col in cat:
    sns.countplot(data=train_iris, x=col)
    plt.title(f'Distribution of {col}')
    plt.show()
    sns.countplot(data=train_iris, x='sepal_length')
    plt.title('Distribution of Sepal Length')
    plt.show()
```

```python
# Document takeaways & any actions.

# there is no significant correlation between species and sepal or petal size
```

```python
# Bivariate Stats

# Visualize each measurement type (y-axis) with the species variable (x-axis) using barplots, adding a horizontal line showing the overall mean of the metric (y-axis).
for col in train_iris:
    if train_iris[col].dtype=='float64':
        quan.append(col)
for col in quan: 
    graph = sns.barplot(data=train_iris, x='species', y=col)
    graph.axhline(train_iris[col].mean())
    plt.title('Sepal and Petal Size by Species')
    plt.show()
```

```python
from scipy import stats
```

```python
# For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) 
# to see if there is a significant difference between the two groups. Do the same for the other measurement types.
virg = train_iris[train_iris.species == 'virginica']
vers = train_iris[train_iris.species == 'versicolor']

t, p = stats.mannwhitneyu(virg.petal_width, vers.petal_width)
t, p
```

```python
if p < .05:
    print('We can reject the null hypothesis')
else:
    print('We cannot reject the null hypotehsis')
```

```python
t, p = stats.mannwhitneyu(virg.petal_length, vers.petal_length)
t, p
```

```python
if p < .05:
    print('We can reject the null hypothesis')
else:
    print('We cannot reject the null hypotehsis')
```

```python
t, p = stats.mannwhitneyu(virg.sepal_width, vers.sepal_width)
t, p
```

```python
if p < .05:
    print('We can reject the null hypothesis')
else:
    print('We cannot reject the null hypotehsis')
```

```python
t, p = stats.mannwhitneyu(virg.sepal_length, vers.sepal_length)
t, p
```

```python
if p < .05:
    print('We can reject the null hypothesis')
else:
    print('We cannot reject the null hypotehsis')
```
